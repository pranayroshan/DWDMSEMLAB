```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from apyori import apriori
```

```python
df = pd.read_csv("groceries - groceries.csv",header = None)
```

```python
df.shape
```

```python
df.head(10)
```

```python
records = []
for i in range(0,9836):
    records.append([str(df.values[i,j]) for j in range(0,33)])
for sets in records:
    print(sets)
```

```python
 A= apriori(records, min_support = 0.06, min_confidence =0.8,min_length = 2)
```

```python
A=list(A)
```

```python
print(len(A))
```

```python
print(A)
```

```python
for item in A:
    pair = item[0]
    items = [x for x in pair]
    print("Rule :"+ str(item[0]) + "->" + str(item[1]))
    print("Support : {}".format(item[1]))
    print("Confidence : {}".format(item[2][0][2]))
    print("List : {}".format(item[2][0][3]))
    print()
    print()
```
